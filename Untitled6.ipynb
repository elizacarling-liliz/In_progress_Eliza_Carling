```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation
import random

class MontyHallAnimation:
    def __init__(self, num_simulations=100):
        self.num_simulations = num_simulations
        self.sim_count = 0
        self.stick_wins = 0
        self.switch_wins = 0

        # Create figure
        self.fig, (self.ax1, self.ax2) = plt.subplots(1, 2, figsize=(12, 5))
        self.fig.suptitle('Monty Hall Problem Simulation', fontsize=16, fontweight='bold')

        # Setup doors
        self.door_colors = ['lightblue', 'lightblue', 'lightblue']
        self.door_texts = ['Door 1', 'Door 2', 'Door 3']
        self.prize_texts = ['?', '?', '?']

        # Car and goats positions (will be set each simulation)
        self.car_door = None
        self.first_choice = None
        self.host_opens = None

        # Statistics
        self.stick_probs = []
        self.switch_probs = []
        self.trials = []

        # Create initial plot
        self.setup_plot()

    def setup_plot(self):
        """Setup the initial plot"""
        # Clear axes
        self.ax1.clear()
        self.ax2.clear()

        # Left plot: Doors
        self.ax1.set_xlim(-0.5, 3.5)
        self.ax1.set_ylim(-0.5, 1.5)
        self.ax1.set_aspect('equal')
        self.ax1.set_title('Current Game', fontsize=14)
        self.ax1.axis('off')

        # Draw doors
        self.door_rects = []
        for i in range(3):
            rect = patches.Rectangle((i, 0), 0.8, 1,
                                    linewidth=2,
                                    edgecolor='black',
                                    facecolor=self.door_colors[i],
                                    alpha=0.8)
            self.ax1.add_patch(rect)
            self.door_rects.append(rect)

            # Door number
            self.ax1.text(i + 0.4, 1.1, f'Door {i+1}',
                         ha='center', va='center', fontsize=12, fontweight='bold')

            # Prize (hidden initially)
            self.ax1.text(i + 0.4, 0.5, '?',
                         ha='center', va='center', fontsize=20, fontweight='bold')

        # Right plot: Statistics
        self.ax2.set_xlim(0, self.num_simulations)
        self.ax2.set_ylim(0, 100)
        self.ax2.set_xlabel('Number of Trials', fontsize=12)
        self.ax2.set_ylabel('Win Percentage (%)', fontsize=12)
        self.ax2.set_title('Win Probabilities Over Time', fontsize=14)
        self.ax2.grid(True, alpha=0.3)

        # Add legend placeholder
        self.ax2.plot([], [], 'b-', linewidth=2, label='Sticking Wins')
        self.ax2.plot([], [], 'r-', linewidth=2, label='Switching Wins')
        self.ax2.legend(loc='lower right')

        # Add text for current stats
        self.stats_text = self.ax2.text(0.02, 0.98, '',
                                       transform=self.ax2.transAxes,
                                       verticalalignment='top',
                                       fontsize=11,
                                       bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

    def run_simulation(self):
        """Run one Monty Hall simulation"""
        # Randomly place car and goats
        self.car_door = random.randint(0, 2)

        # Contestant's first choice
        self.first_choice = random.randint(0, 2)

        # Host opens a door with a goat (not contestant's choice)
        available_doors = [0, 1, 2]
        available_doors.remove(self.first_choice)

        if self.car_door in available_doors:
            available_doors.remove(self.car_door)

        self.host_opens = random.choice(available_doors)

        # Determine switch door
        remaining_doors = [0, 1, 2]
        remaining_doors.remove(self.first_choice)
        remaining_doors.remove(self.host_opens)
        switch_door = remaining_doors[0]

        # Update wins
        self.sim_count += 1
        if self.first_choice == self.car_door:
            self.stick_wins += 1
        if switch_door == self.car_door:
            self.switch_wins += 1

        # Update probabilities
        stick_prob = (self.stick_wins / self.sim_count) * 100
        switch_prob = (self.switch_wins / self.sim_count) * 100

        self.stick_probs.append(stick_prob)
        self.switch_probs.append(switch_prob)
        self.trials.append(self.sim_count)

        return stick_prob, switch_prob

    def update_plot(self, frame):
        """Update the animation frame"""
        # Run one simulation
        stick_prob, switch_prob = self.run_simulation()

        # Clear and redraw left plot (doors)
        self.ax1.clear()
        self.ax1.set_xlim(-0.5, 3.5)
        self.ax1.set_ylim(-0.5, 1.5)
        self.ax1.set_aspect('equal')
        self.ax1.set_title(f'Game #{self.sim_count}', fontsize=14)
        self.ax1.axis('off')

        # Draw doors with appropriate colors
        for i in range(3):
            # Determine door color
            if i == self.car_door:
                door_color = 'gold'  # Car is behind this door
                prize_text = ' CAR!'
            elif i == self.host_opens:
                door_color = 'lightgray'  # Opened door
                prize_text = ' Goat'
            else:
                door_color = 'lightblue'  # Closed door
                prize_text = '?'

            # Draw door rectangle
            rect = patches.Rectangle((i, 0), 0.8, 1,
                                    linewidth=3 if i == self.first_choice else 2,
                                    edgecolor='red' if i == self.first_choice else 'black',
                                    facecolor=door_color,
                                    alpha=0.8)
            self.ax1.add_patch(rect)

            # Door number
            self.ax1.text(i + 0.4, 1.1, f'Door {i+1}',
                         ha='center', va='center', fontsize=12, fontweight='bold')

            # What's behind the door (revealed for opened door and car at end)
            if i == self.host_opens or (frame == self.num_simulations - 1):
                self.ax1.text(i + 0.4, 0.5, prize_text,
                             ha='center', va='center', fontsize=14, fontweight='bold')
            else:
                self.ax1.text(i + 0.4, 0.5, '?',
                             ha='center', va='center', fontsize=20, fontweight='bold')

        # Add text annotations
        info_text = f"Contestant chose: Door {self.first_choice + 1}\n"
        info_text += f"Host opened: Door {self.host_opens + 1} (Goat)\n"
        info_text += f"Car is behind: Door {self.car_door + 1}"

        self.ax1.text(1.5, -0.2, info_text,
                     ha='center', va='top', fontsize=10,
                     bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

        # Update right plot (statistics)
        self.ax2.clear()
        self.ax2.set_xlim(0, max(10, self.sim_count))
        self.ax2.set_ylim(0, 100)
        self.ax2.set_xlabel('Number of Trials', fontsize=12)
        self.ax2.set_ylabel('Win Percentage (%)', fontsize=12)
        self.ax2.set_title(f'Win Probabilities Over Time\n(Trial {self.sim_count}/{self.num_simulations})', fontsize=14)
        self.ax2.grid(True, alpha=0.3)

        # Plot probability curves
        if len(self.trials) > 0:
            self.ax2.plot(self.trials, self.stick_probs, 'b-', linewidth=2, label='Sticking Wins')
            self.ax2.plot(self.trials, self.switch_probs, 'r-', linewidth=2, label='Switching Wins')

            # Add horizontal lines at theoretical values
            self.ax2.axhline(y=33.33, color='b', linestyle='--', alpha=0.3, linewidth=1)
            self.ax2.axhline(y=66.67, color='r', linestyle='--', alpha=0.3, linewidth=1)

        self.ax2.legend(loc='lower right')

        # Update statistics text
        stats_text = f"Current Trial: {self.sim_count}\n"
        stats_text += f"Sticking Wins: {self.stick_wins} ({stick_prob:.1f}%)\n"
        stats_text += f"Switching Wins: {self.switch_wins} ({switch_prob:.1f}%)\n"
        stats_text += f"Theoretical: 33.3% vs 66.7%"

        self.ax2.text(0.02, 0.98, stats_text,
                     transform=self.ax2.transAxes,
                     verticalalignment='top',
                     fontsize=11,
                     bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

        # Add final message if done
        if frame == self.num_simulations - 1:
            conclusion = "CONCLUSION: Switching doubles your chance of winning!"
            self.fig.text(0.5, 0.02, conclusion,
                         ha='center', va='bottom', fontsize=14,
                         fontweight='bold', color='darkred',
                         bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9))

        return self.ax1, self.ax2

    def animate(self):
        """Create the animation"""
        anim = animation.FuncAnimation(self.fig, self.update_plot,
                                      frames=self.num_simulations,
                                      interval=500,  # 0.5 seconds between frames
                                      repeat=False)
        return anim

def create_monty_hall_video(num_simulations=50, save_as_gif=False, filename='monty_hall.gif'):
    """
    Create and save a Monty Hall animation

    Parameters:
    num_simulations: Number of games to simulate
    save_as_gif: If True, save as GIF file
    filename: Name of output file
    """
    print("Creating Monty Hall animation...")
    print(f"Running {num_simulations} simulations...")

    # Create animation
    mh = MontyHallAnimation(num_simulations)
    anim = mh.animate()

    # Save as GIF if requested
    if save_as_gif:
        print(f"Saving animation to {filename}...")
        anim.save(filename, writer='pillow', fps=2)
        print(f"Animation saved to {filename}")

    # Show the animation
    plt.tight_layout()
    plt.show()

    return anim

# Run the animation
if __name__ == "__main__":
    # For quick testing: 20 simulations (shows the trend clearly)
    anim = create_monty_hall_video(num_simulations=30, save_as_gif=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
import random
import time

# Set up the figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('MONTY HALL PROBLEM: Should You Switch Doors?', fontsize=18, fontweight='bold', y=0.98)

# Animation control
current_step = 0
total_simulations = 30
simulation_count = 0
stick_wins = 0
switch_wins = 0
stick_probs = []
switch_probs = []
trials = []

# Game state
car_door = None
first_choice = None
host_opens = None
revealed = False
game_state = "initial"  # initial, chosen, revealed, final

def setup_initial_doors():
    """Draw the initial 3 doors"""
    ax1.clear()
    ax1.set_xlim(-1, 4)
    ax1.set_ylim(-1, 3)
    ax1.set_aspect('equal')
    ax1.axis('off')
    ax1.set_title('DOORS', fontsize=14, fontweight='bold', pad=10)

    # Draw 3 doors
    door_rects = []
    for i in range(3):
        # Door rectangle
        rect = patches.Rectangle((i, 0), 0.8, 2,
                                linewidth=3,
                                edgecolor='darkblue',
                                facecolor='lightblue',
                                alpha=0.9)
        ax1.add_patch(rect)
        door_rects.append(rect)

        # Door number
        ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                ha='center', va='center',
                fontsize=14, fontweight='bold', color='darkblue')

        # Question mark inside
        ax1.text(i + 0.4, 1, '?',
                ha='center', va='center',
                fontsize=36, fontweight='bold', color='gray')

    # Instructions
    ax1.text(1.5, -0.5, 'Pick a door!',
             ha='center', va='center',
             fontsize=12, style='italic', color='darkgreen')

    return door_rects

def setup_statistics():
    """Setup the statistics plot"""
    ax2.clear()
    ax2.set_xlim(0, total_simulations)
    ax2.set_ylim(0, 100)
    ax2.set_xlabel('Number of Games', fontsize=12)
    ax2.set_ylabel('Win Percentage (%)', fontsize=12)
    ax2.set_title('WINNING PROBABILITIES', fontsize=14, fontweight='bold', pad=10)
    ax2.grid(True, alpha=0.3, linestyle='--')

    # Theoretical lines
    ax2.axhline(y=33.33, color='blue', linestyle=':', alpha=0.5, linewidth=2, label='Theoretical: Sticking (33.3%)')
    ax2.axhline(y=66.67, color='red', linestyle=':', alpha=0.5, linewidth=2, label='Theoretical: Switching (66.7%)')

    return ax2

def create_new_game():
    """Create a new Monty Hall game"""
    global car_door, first_choice, host_opens, revealed, game_state
    global simulation_count

    simulation_count += 1

    # Place car randomly
    car_door = random.randint(0, 2)

    # Contestant picks randomly
    first_choice = random.randint(0, 2)

    # Host opens a door with a goat
    available_doors = [0, 1, 2]
    available_doors.remove(first_choice)

    if car_door in available_doors:
        available_doors.remove(car_door)

    host_opens = random.choice(available_doors)

    # Reset state
    revealed = False
    game_state = "initial"

    return car_door, first_choice, host_opens

def draw_game_state(step):
    """Draw the current game state based on animation step"""
    global revealed, game_state
    global stick_wins, switch_wins

    ax1.clear()
    ax1.set_xlim(-1, 4)
    ax1.set_ylim(-1, 3)
    ax1.set_aspect('equal')
    ax1.axis('off')

    # Determine which step we're in for this game
    game_progress = step % 4  # 0,1,2,3 for the 4 phases of each game

    if game_progress == 0:  # Show initial doors
        game_state = "initial"
        ax1.set_title('Step 1: Initial Doors', fontsize=14, fontweight='bold', pad=10)

        for i in range(3):
            rect = patches.Rectangle((i, 0), 0.8, 2,
                                    linewidth=3,
                                    edgecolor='darkblue',
                                    facecolor='lightblue')
            ax1.add_patch(rect)
            ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                    ha='center', va='center', fontsize=14, fontweight='bold')
            ax1.text(i + 0.4, 1, '?',
                    ha='center', va='center', fontsize=36, fontweight='bold', color='gray')

        ax1.text(1.5, -0.5, f'Game #{simulation_count}: Car is randomly placed behind one door',
                ha='center', va='center', fontsize=11, color='darkgreen')

    elif game_progress == 1:  # Show contestant's choice
        game_state = "chosen"
        ax1.set_title('Step 2: You Choose a Door', fontsize=14, fontweight='bold', pad=10)

        for i in range(3):
            # Highlight chosen door
            if i == first_choice:
                edge_color = 'red'
                edge_width = 5
                face_color = 'lightcoral'
            else:
                edge_color = 'darkblue'
                edge_width = 3
                face_color = 'lightblue'

            rect = patches.Rectangle((i, 0), 0.8, 2,
                                    linewidth=edge_width,
                                    edgecolor=edge_color,
                                    facecolor=face_color)
            ax1.add_patch(rect)
            ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                    ha='center', va='center', fontsize=14, fontweight='bold')
            ax1.text(i + 0.4, 1, '?',
                    ha='center', va='center', fontsize=36, fontweight='bold', color='gray')

        ax1.text(1.5, -0.5, f'You chose Door {first_choice + 1}',
                ha='center', va='center', fontsize=11, color='red', fontweight='bold')

    elif game_progress == 2:  # Show host opening a door
        game_state = "revealed"
        ax1.set_title('Step 3: Host Reveals a Goat', fontsize=14, fontweight='bold', pad=10)

        for i in range(3):
            if i == host_opens:  # Opened door
                # Draw open door (show goat)
                rect = patches.Rectangle((i, 0), 0.8, 2,
                                        linewidth=3,
                                        edgecolor='gray',
                                        facecolor='lightgray',
                                        alpha=0.7)
                ax1.add_patch(rect)
                ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                        ha='center', va='center', fontsize=14, fontweight='bold', color='gray')
                ax1.text(i + 0.4, 1, '',
                        ha='center', va='center', fontsize=36)

                # Add "OPENED" text
                ax1.text(i + 0.4, -0.3, 'OPENED',
                        ha='center', va='center', fontsize=10, color='gray', fontweight='bold')

            elif i == first_choice:  # Your chosen door
                edge_color = 'red'
                edge_width = 5
                face_color = 'lightcoral'
                rect = patches.Rectangle((i, 0), 0.8, 2,
                                        linewidth=edge_width,
                                        edgecolor=edge_color,
                                        facecolor=face_color)
                ax1.add_patch(rect)
                ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                        ha='center', va='center', fontsize=14, fontweight='bold')
                ax1.text(i + 0.4, 1, '?',
                        ha='center', va='center', fontsize=36, fontweight='bold', color='gray')
            else:  # Other closed door
                rect = patches.Rectangle((i, 0), 0.8, 2,
                                        linewidth=3,
                                        edgecolor='darkblue',
                                        facecolor='lightblue')
                ax1.add_patch(rect)
                ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                        ha='center', va='center', fontsize=14, fontweight='bold')
                ax1.text(i + 0.4, 1, '?',
                        ha='center', va='center', fontsize=36, fontweight='bold', color='gray')

        ax1.text(1.5, -0.5, f'Host opens Door {host_opens + 1} to reveal a goat',
                ha='center', va='center', fontsize=11, color='purple', fontweight='bold')
        ax1.text(1.5, -0.8, 'You can STICK or SWITCH to the other unopened door',
                ha='center', va='center', fontsize=10, style='italic')

    elif game_progress == 3:  # Final reveal
        game_state = "final"
        revealed = True
        ax1.set_title('Step 4: Final Reveal!', fontsize=14, fontweight='bold', pad=10)

        # Calculate if sticking or switching would win
        stick_wins_this = (first_choice == car_door)
        remaining_doors = [0, 1, 2]
        remaining_doors.remove(first_choice)
        remaining_doors.remove(host_opens)
        switch_choice = remaining_doors[0]
        switch_wins_this = (switch_choice == car_door)

        if stick_wins_this:
            stick_wins += 1
        if switch_wins_this:
            switch_wins += 1

        # Update probabilities
        stick_prob = (stick_wins / simulation_count) * 100
        switch_prob = (switch_wins / simulation_count) * 100
        stick_probs.append(stick_prob)
        switch_probs.append(switch_prob)
        trials.append(simulation_count)

        # Draw all doors with what's behind them
        for i in range(3):
            if i == car_door:  # Door with car
                rect = patches.Rectangle((i, 0), 0.8, 2,
                                        linewidth=5,
                                        edgecolor='gold',
                                        facecolor='lightyellow')
                ax1.add_patch(rect)
                ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                        ha='center', va='center', fontsize=14, fontweight='bold', color='goldenrod')
                ax1.text(i + 0.4, 1, ' WIN!',
                        ha='center', va='center', fontsize=20, fontweight='bold', color='darkorange')

            elif i == host_opens:  # Opened door (goat)
                rect = patches.Rectangle((i, 0), 0.8, 2,
                                        linewidth=3,
                                        edgecolor='gray',
                                        facecolor='lightgray')
                ax1.add_patch(rect)
                ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                        ha='center', va='center', fontsize=14, color='gray')
                ax1.text(i + 0.4, 1, ' Goat',
                        ha='center', va='center', fontsize=16, color='gray')
            else:  # Other door (goat)
                rect = patches.Rectangle((i, 0), 0.8, 2,
                                        linewidth=3,
                                        edgecolor='darkblue',
                                        facecolor='lightblue')
                ax1.add_patch(rect)
                ax1.text(i + 0.4, 2.2, f'DOOR {i+1}',
                        ha='center', va='center', fontsize=14)
                ax1.text(i + 0.4, 1, ' Goat',
                        ha='center', va='center', fontsize=16, color='gray')

            # Highlight your initial choice
            if i == first_choice:
                # Add a star above the chosen door
                ax1.plot([i + 0.4], [2.4], 'r*', markersize=15)
                ax1.text(i + 0.4, 2.6, 'Your pick',
                        ha='center', va='center', fontsize=10, color='red', fontweight='bold')

        # Show result text
        result_text = f"Car was behind Door {car_door + 1}\n"
        result_text += f"You picked Door {first_choice + 1}\n"

        if stick_wins_this:
            result_text += "STICKING would WIN!\n"
            result_color = 'blue'
        else:
            result_text += "STICKING would LOSE\n"
            result_color = 'gray'

        result_text += "\n"

        if switch_wins_this:
            result_text += "SWITCHING would WIN!"
            result_color = 'red' if not stick_wins_this else 'green'
        else:
            result_text += "SWITCHING would LOSE"

        ax1.text(1.5, -0.5, result_text,
                ha='center', va='center', fontsize=11,
                color=result_color, fontweight='bold',
                bbox=dict(boxstyle='round', facecolor='whitesmoke', alpha=0.9))

    return ax1

def update_stats():
    """Update the statistics plot"""
    ax2.clear()
    ax2.set_xlim(0, max(10, simulation_count + 1))
    ax2.set_ylim(0, 100)
    ax2.set_xlabel('Number of Games', fontsize=12)
    ax2.set_ylabel('Win Percentage (%)', fontsize=12)
    ax2.set_title(f'STATISTICS (Game {simulation_count}/{total_simulations})',
                 fontsize=14, fontweight='bold', pad=10)
    ax2.grid(True, alpha=0.3, linestyle='--')

    # Plot probability curves if we have data
    if len(trials) > 0:
        ax2.plot(trials, stick_probs, 'b-', linewidth=3, label=f'Sticking ({stick_probs[-1]:.1f}%)')
        ax2.plot(trials, switch_probs, 'r-', linewidth=3, label=f'Switching ({switch_probs[-1]:.1f}%)')

        # Theoretical lines
        ax2.axhline(y=33.33, color='blue', linestyle=':', alpha=0.5, linewidth=2)
        ax2.axhline(y=66.67, color='red', linestyle=':', alpha=0.5, linewidth=2)

        # Add current point markers
        ax2.plot(trials[-1], stick_probs[-1], 'bo', markersize=8)
        ax2.plot(trials[-1], switch_probs[-1], 'ro', markersize=8)

    ax2.legend(loc='lower right', fontsize=11)

    # Add statistics text box
    if simulation_count > 0:
        stats_text = f"Games Played: {simulation_count}\n"
        stats_text += f"Sticking Wins: {stick_wins} ({stick_probs[-1]:.1f}%)\n"
        stats_text += f"Switching Wins: {switch_wins} ({switch_probs[-1]:.1f}%)\n"
        stats_text += f"\nTheoretical:\n"
        stats_text += f"Sticking: 33.3%\n"
        stats_text += f"Switching: 66.7%"

        ax2.text(0.02, 0.98, stats_text,
                transform=ax2.transAxes,
                verticalalignment='top',
                fontsize=11,
                bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

    return ax2

def animate(frame):
    """Main animation function"""
    global current_step, car_door, first_choice, host_opens

    # Clear both axes
    ax1.clear()
    ax2.clear()

    # Calculate which game and which step we're on
    game_num = frame // 4  # Each game has 4 steps
    step_in_game = frame % 4

    # Create new game if needed
    if step_in_game == 0:
        car_door, first_choice, host_opens = create_new_game()

    # Draw the current game state
    draw_game_state(frame)

    # Update statistics (but only after games are complete)
    if step_in_game == 3 and game_num > 0:
        update_stats()
    else:
        update_stats()

    # Add overall progress
    progress = (frame + 1) / (total_simulations * 4)
    fig.text(0.5, 0.01, f'Progress: {progress*100:.1f}%  Press SPACE to pause/resume',
            ha='center', va='bottom', fontsize=10, style='italic')

    # Check if we're done
    if frame >= total_simulations * 4 - 1:
        # Add final conclusion
        conclusion = "\n" + "="*60 + "\n"
        conclusion += "FINAL CONCLUSION: After many simulations...\n"
        conclusion += "SWITCHING DOORS DOUBLES YOUR CHANCE OF WINNING!\n"
        conclusion += f"Sticking wins ~{stick_probs[-1]:.1f}% vs Switching wins ~{switch_probs[-1]:.1f}%\n"
        conclusion += "="*60

        fig.text(0.5, 0.05, conclusion,
                ha='center', va='bottom',
                fontsize=12, fontweight='bold',
                color='darkred',
                bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9))

    return ax1, ax2

# Setup initial plots
setup_initial_doors()
setup_statistics()

# Create the animation
print("Creating Monty Hall animation...")
print(f"Each game has 4 steps, total {total_simulations} games")
print("Animation will show step-by-step...")

# Create animation
anim = animation.FuncAnimation(fig, animate,
                              frames=total_simulations * 4,  # 4 steps per game
                              interval=800,  # 0.8 seconds per frame
                              repeat=False)

# Save as GIF
print("\nSaving animation as 'monty_hall_animated.gif'...")
anim.save('monty_hall_animated.gif', writer=PillowWriter(fps=1.5))  # 1.5 frames per second
print("Animation saved successfully!")

# Also display it
plt.tight_layout()
plt.show()

# Print final statistics
print("\n" + "="*60)
print("FINAL RESULTS:")
print(f"Total games simulated: {simulation_count}")
print(f"Sticking wins: {stick_wins} ({stick_wins/simulation_count*100:.1f}%)")
print(f"Switching wins: {switch_wins} ({switch_wins/simulation_count*100:.1f}%)")
print("="*60)
```
