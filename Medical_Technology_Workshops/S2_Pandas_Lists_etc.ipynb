Before starting today's lecture, let's quickly sidetrack and cover what a **function** is. We've been using them throughout the course, but it's worth seeing how they're structured and how we can create our own functions.

A function takes the given inputs, and uses them to get to a desired output. They are used extensively in python to avoid repeating sections of code.

Below is a function called **_adder_** that takes in two numbers and returns the sum of the two numbers.

A function in Python is defined by using the **def** keyword, followed by a space and the name of the function.

Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, separating them with a comma.

Arguments with an equal sign defines default values.

Our _adder_ function has two arguments (n1, n2). When the function is called, we pass along the first number we want to add and then the second number after the comma, which is then used inside the function to return the sum of the arguments.


```python
def adder(n1, n2):
  total = n1 + n2
  return total

sum_of_numbers = adder(2, 3)
print(sum_of_numbers)

#Â alternatively, we could've written the two lines above in a single line as shown below

print(adder(2, 3))
```

```python
def adder(n1, n2=1):
  '''
  This function adds 2 numbers together. This function adds 1 by default
  '''
  total = n1 + n2
  return total

print(adder(2)) # no n2 specified
print(adder(2,3)) # the function works as before
```

## Introduction to NumPy and NumPy Arrays

While lists are very versatile data structures that can be used in Python, there are times when we need to analyse vast quantities of data and perform mathematical operations on them. We are going to use the NumPy ("Numerical Python") library to help us with this.

In order to access NumPy and its functionalities, we do the following:

```python
import numpy as np
```

To convert a list into a NumPy array, we simply do the following:     

```python
heights = [190, 160, 170, 180, 176]

heights_array = np.array(heights)

heights_array + 1
```

We are now going to look at calculating the body mass index (BMI) for our group of people. To do this we will need to take their masses and divide by their height squared.

Notice that our array of heights is in centimetres. However, we need this array to be in metres. The beauty of numpy arrays is that we can perform mathematical operations in a single line of code. For example, to calculate the height of each member of the group in metres, we can simply divide the array by 100 and it will perform the calculation for each element of the array. For example:



```python
heights_metres = heights_array/100
print(heights_metres)
```

We can also perform operations between arrays. There are some caveats that depend on the shapes of the arrays, but these are beyond the scope of what will be discussed today. A basic example:

```python
distance_run_day1 = np.array([5, 10, 23, 11, 4])
distance_run_day2 = np.array([10, 20, 1, 0, 4])
print(distance_run_day1 + distance_run_day2)
```

We can also multiply, divide or square a numpy array and that operation will be applied to each element.

## Introduction to Matplotlib

Matplotlib is a library for visualising data in Python. Using Matplotlib we can plot a wide variety of graphs to help us visualise our data. We first start off by **importing** a specific part of matplotlib (pyplot) into our **namespace**.

```python
import matplotlib.pyplot as plt
```

There are a variety of plots that can be done in matplotlib. We will only consider two in this session: scatter/line plots and histogram. Scatter plots can be created using the plt.scatter() function, e.g.:      

x = [1, 2, 3, 4]

y = [2, 4, 6, 8]

plt.scatter(x, y)

You can label your x and y axes by using the plt.xlabel() and plt.ylabel() functions. Simply make sure to provide the text you want as the label inside the brackets within quotation marks.

plt.xlabel("x")

plt.ylabel("y")

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("This is a scatter plot")
plt.show()
```

You can even plot multiple scatter plots on the same graph! In order to be able to distinguish them, you can add an extra argument inside the brackets called "label", which allows you to provide a name for each scatter plot. You then need to add plt.legend() afterwards to create a legend that will give you a colour code!

```python
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [3, 6, 9, 12]
plt.scatter(x, y1, label = "A")
plt.scatter(x, y2, label = "B")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

We're now going to consider histograms which tell us how our data is distributed.

Below I generate 1000 random values which I assign to the variable **x**.

```python
mu, sigma = 20, 3
x = mu + sigma * np.random.randn(1000)
plt.hist(x, density=True, histtype = "step")
plt.xlabel("y")
plt.ylabel("Density")
plt.show()
```
